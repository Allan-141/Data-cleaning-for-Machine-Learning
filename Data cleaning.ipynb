## Removing unwanted values ie: N/a,NaN,Na,np.nan

### I use dropna() function drop Null values

```python
df.dropna()
```

### I can also set to drop rows that are completly null only

### Let me first create a simple data set to explain.

```python
df1= pd.DataFrame(data=
                 {
                     "TEMPERATURE": [36,34,np.nan,35,37,np.nan,38,39],
                     "BLOOD PRESSURE": [86,np.nan,78,100,84,np.nan,90,77]
                 }
                 )
```

```python
df1
```

#### Drops all of row 5 because it has all its columns NaN

```python
df1.dropna(how="all")
```

#### Drops all the rows with any NaN value(s)

```python
df1.dropna()
```

### Fills all the null values ,I do it with zeros

```python
df.fillna(0)
```

## But rather than filling all the null values with zeros I can use forward fill method='ffill'

### This fills the empty values with the value above it

```python
df.fillna(method = 'ffill')
```

## I can also rather fill all the null values with backward fill method='bfill'

```python
df.fillna(method = 'bfill')
```

## On the other dataset I had  created looks like this:

```python
df1
```

```python
df1.fillna(method="ffill")
```

```python
df1.fillna(method = 'bfill')
```

## The best method that I recommend on filling the null values is by INTERPOLATION eg:(34+35)2 to get row 2 

```python
df1.interpolate()
```
